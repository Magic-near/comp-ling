```python
import re
import json
```

```python
with open('corpus.json', 'r') as f:
    data = json.loads(f.readline())
```

```python
for post in range(5):
    print(data[str(post)]['post_text'])
    print('')
    print('**************************************************************************')
    print('')
```

```python
pattern = r'\n'
for post in data:
    data[post]['post_text'] = re.sub(pattern, ' ', data[post]['post_text'])
```

```python
pattern = r'#\w+@\w+'
for post in data:
    data[post]['post_text'] = re.sub(pattern, '', data[post]['post_text'])
```

```python
pattern = r'#\w+'
for post in data:
    data[post]['post_text'] = re.sub(pattern, '', data[post]['post_text'])
```

```python
pattern = r'http\S+'
for post in data:
    data[post]['post_text'] = re.sub(pattern, '', data[post]['post_text'], flags=re.MULTILINE)
```

```python
pattern = r'[^a-zA-Z–∞-—è—ë–ê-–Ø–Å\s\d!"#$%&/\\\\\'\(\)\*\+,\-\.:;<=>\?@\[\]^_`{|}~]+'
for post in data:
    data[post]['post_text'] = re.sub(pattern, '', data[post]['post_text'])
```

```python
pattern = r'id\d+'
for post in data:
    data[post]['post_text'] = re.sub(pattern, '', data[post]['post_text'])
```

```python
pattern = r'club\d+'
for post in data:
    data[post]['post_text'] = re.sub(pattern, '', data[post]['post_text'])
```

```python
pattern = r'\|'
for post in data:
    data[post]['post_text'] = re.sub(pattern, '', data[post]['post_text'])
```

```python
pattern = r'\s+'
for post in data:
    data[post]['post_text'] = re.sub(pattern, ' ', data[post]['post_text'])
```

```python
for post in range(5):
    print(data[str(post)]['post_text'])
    print('')
    print('**************************************************************************')
    print('')
```

```python
with open("clean_corpus.json", "w") as f:
    json.dump(data, f, ensure_ascii=False)
```
